# Descrição do Problema

O conjunto de dados **Dados históricos das Olimpíadas** é um conjunto de microdados de atletas e delegações desde 1896 até 2016, originalmente consolidade no Kaggle e disponibilizada na basedosdados.org.

Os Jogos Olímpicos reúnem atletas de diferentes países, modalidades esportivas e perfis físicos. ## Análise de valores faltantes

Para cada atributo do conjunto de dados, vou verificar quantas entradas estão com valores faltantes e qual o percentual que esse resultado representa no total de entradas do atributo e no total de entradas faltantes em todo conjunto de dados. Gerei uma visualização em tabela e dois gráficos de barras horizontais com os resultados.

````python
total_missing_values = df.isna().sum().sum()

missing_por_coluna = (
    df.isna().sum().to_frame("qtd_missing")
    .assign(
        pct_missing=lambda x: (x["qtd_missing"] / len(df) * 100).round(1),
        pct_total_missing=lambda x: (x["qtd_missing"] / total_missing_values * 100).round(1)
    )
    .sort_values(["pct_missing","qtd_missing"],ascending=False)
)
missing_por_coluna.head(15)
````

````python
# Percentual de Valores Faltantes por Coluna
plt.figure(figsize=(8, 5))
missing_por_coluna_filtrado = missing_por_coluna[missing_por_coluna['pct_missing'] !=0]
bars = plt.barh(missing_por_coluna_filtrado.index, missing_por_coluna_filtrado['pct_missing'], color='skyblue')
plt.title("Percentual de Valores Faltantes por Coluna")
plt.xlabel("Percentual (%)")

for bar in bars:
    width = bar.get_width()
    plt.text(width, bar.get_y() + bar.get_height()/2, f'{width:.1f}%', va='center')

plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
````

````python
# Percentual de Cada Coluna no Total de Valores Faltantes do DataFrame
plt.figure(figsize=(8, 5))
missing_por_coluna_filtrado2 = missing_por_coluna[missing_por_coluna['pct_total_missing'] !=0]
bars = plt.barh(missing_por_coluna_filtrado2.index, missing_por_coluna_filtrado2['pct_total_missing'], color='blue')
plt.title("Percentual de Cada Coluna no Total de Valores Faltantes do DataFrame")
plt.xlabel("Percentual (%)")

for bar in bars:
    width = bar.get_width()
    plt.text(width, bar.get_y() + bar.get_height()/2, f'{width:.1f}%', va='center')

plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
````

O atributo "medalha" concentra a maior parte das células faltantes. Isso faz sentido pois, em cada esporte/evento, muitos atletas participam e apenas três atletas ou equipes de atletas recebem medalhas pela classificação em primeiro (ouro), segundo (prata) ou terceiro (bronze) lugares. ### Média e Desvio Padrão

A média é uma medida de tendência central que representa o valor típico ou o ponto de equilíbrio de um conjunto de dados. É calculada somando-se todos os valores e dividindo-se pelo número total de observações. É sensível a valores extremos (outliers).

O desvio padrão é uma medida de dispersão que quantifica a quantidade de variação ou dispersão de um conjunto de valores. Um desvio padrão baixo indica que os pontos de dados tendem a estar próximos da média do conjunto, enquanto um desvio padrão alto indica que os pontos de dados estão espalhados por uma faixa maior de valores. Ele é a raiz quadrada da variância.

````python
# média dos atributos numéricos (altura, peso e idade) (separei aqui os que fazem sentido calcular média)
cols = ['idade', 'peso', 'altura']
df[cols].describe().loc['mean'].round(1)
````

````python
# desvio padrão dos atributos numéricos (altura, peso e idade) (separei aqui os que fazem sentido calcular desvio padrão)
df[cols].describe().loc['std'].round(1)
````

````python
# Qual gráfico seria interessante usar pra mostrar a média? Qual gráfico seria interessante usar pra mostrar o desvio padrão?
````

Para visualizar a média e o desvio padrão dos atributos "idade", "peso" e "altura", optei por utilizar o mesmo tipo de gráfico (histograma).

````python
df_plot = df[cols].apply(pd.to_numeric, errors='coerce')

fig, axes = plt.subplots(1, len(cols), figsize=(18, 5))

for ax, col in zip(axes, cols):
    # Série sem NaN
    s = df_plot[col].dropna()

    # Histograma
    sns.histplot(s, bins=30,ax=ax)

    # Média e desvio padrão
    mean_val = s.mean()
    std_val = s.std()
    ax.axvline(mean_val, color='red', linestyle='--', linewidth=2, label=f"Média = {mean_val:.1f}")
    ax.axvline(mean_val+std_val, color='gray', linestyle='--', linewidth=2, label=f"Desvio = {std_val:.1f}")
    ax.axvline(mean_val-std_val, color='gray', linestyle='--', linewidth=2)

    # Títulos e legenda
    ax.set_title(f"Histograma de {col}")
    ax.set_xlabel(col)
    ax.set_ylabel("Frequência")
    ax.legend()

plt.tight_layout()
plt.show()
````

No próximo histograma, irei plotar a curva kde do peso considerando os 5 esportes mais comuns nas Olimpiadas.

````python
top_n = 5
top_esportes = df['esporte'].value_counts().head(top_n).index
df_top = df[df['esporte'].isin(top_esportes)].copy()

plt.figure(figsize=(10, 6))
for esp in top_esportes:
    s = df_top.loc[df_top['esporte'] == esp, 'peso'].dropna()
    sns.kdeplot(s, label=esp)

plt.axvline(45, color='gray', linestyle='--', linewidth=1, label='45 kg')
plt.axvline(100, color='gray', linestyle='--', linewidth=1, label='100 kg')
plt.title(f'Curvas KDE de peso — comparação entre esportes (Top {top_n})')
plt.xlabel('Peso (kg)')
plt.ylabel('Densidade')
plt.legend()
plt.show()
````

Como o perfil dos atletas pode variar bastante entre os esportes, modalidades do esporte e sexo, vou analisar alguns histogramas agregando as informações de peso (e posteriormente de altura) segundo os esportes, modalidades praticadas e sexo do atleta.

````python
# Parâmetros
top_sports_n = 3
top_events_n = 6
col_sport = "esporte"
col_event = "evento"
col_value = "peso"
col_sex   = "sexo"

# Parâmetros dos gráficos
ncols = 2
nrows = top_sports_n
fig_width  = ncols * 8.5
fig_height = nrows * 4.5

fig, axes = plt.subplots(
    nrows, ncols,
    figsize=(fig_width, fig_height),
    constrained_layout=True
)

axes = axes.flatten()

# Preparação dos dados
df_plot = df.copy()
df_plot[col_value] = pd.to_numeric(df_plot[col_value], errors="coerce")
df_plot = df_plot.dropna(subset=[col_sport, col_event, col_value, col_sex])

# Top esportes
top_sports = (
        df_plot[col_sport]
        .value_counts()
        .head(top_sports_n)
        .index
        .tolist()
    )

def plot_kde_por_sexo(sex_value, sex_label, col_index):
    df_sex = df_plot[df_plot[col_sex] == sex_value].copy()



    for i, sport in enumerate(top_sports):
        ax = axes[i * ncols + col_index]
        df_sport = df_sex[df_sex[col_sport] == sport].copy()

        # Top eventos dentro do esporte
        top_events = (
            df_sport[col_event]
            .value_counts()
            .head(top_events_n)
            .index
            .tolist()
        )

        df_sport = df_sport[df_sport[col_event].isin(top_events)]

        # KDE por evento
        for ev in top_events:
            s = df_sport.loc[df_sport[col_event] == ev, col_value]
            sns.kdeplot(
                s,
                ax=ax,
                linewidth=1.8,
                label=ev
              )

        ax.set_title(f"{sport} — {sex_label}", fontsize=11, pad=10)
        ax.set_xlabel("Peso (kg)")
        ax.set_ylabel("Densidade")

        ax.legend(
            fontsize=6,
            loc="upper right"
        )


plot_kde_por_sexo("M", "Homens", col_index=0)
plot_kde_por_sexo("F", "Mulheres", col_index=1)

plt.suptitle(
      f"Distribuição do peso (KDE) | Top {top_sports_n} esportes e Top {top_events_n} eventos",
      fontsize=14
  )
plt.show()
````

Pelos histogramas anteriores, fica claro que o peso do atleta, além de variar entre os esportes, varia também entre as modalidades do esporte. Fiz o histograma para os 3 esportes mais comuns e coloquei as 6 modalidades mais comuns. Também dividi entre homens e mulheres. No atletismo (modalidade de maratona masculina), o peso dos atletas apresenta tendência a valores mais baixos que em corridas de menor distância. O mesmo ocorre para as mulheres. Na natação e ginástica, observei separação das distribuições em relação a mulheres (menor peso) e homens (maior peso), mas com pouca variação entre as modalidades avaliadas.

## Matriz de Correlação

A matriz de correlação mede a força e a direção de uma relação linear que os atributos numéricos das espécies podem ter. Valores próximos a 1 indicam uma forte correlação positiva, -1 uma forte correlação negativa, e 0 ausência de correlação linear.

Elaborei acima os mapas de calor das matrizes de correlação usando dois métodos (Pearson e Spearman). A correlação de Pearson mede a relação linear entre duas variáveis contínuas, assumindo que os dados seguem uma distribuição normal e a relação entre as variáveis é linear. A correlação de Spearman mede a relação monotônica entre duas variáveis, sendo menos sensível a outliers.
Existe uma forte correlação positiva entre peso e altura (quanto maior a altura, maior tende a ser o peso do atleta). Os dois métodos apresentam valores altos e semelhantes, indicando que o peso e altura do atleta estão fortemente associados no conjunto de dados estudados.
A correlação entre idade e peso (ou altura) é fraca, indicando que a idade do atleta tem pouca influência sobre o peso (ou altura).
Existe uma correlação moderada e positiva entre sexo e peso (ou altura). Como a codificação de "sexo_bin" foi para Masculino=1, os valores positivos obtidos para correlação indicam que os atletas homens tendem a ter peso e altura maior que as atletas mulheres. # Pré-Processamento de Dados

O pré-processamento de dados é uma etapa crucial para preparar os dados para modelagem, garantindo que estejam no formato correto e otimizados para o desempenho do algoritmo.

**Separar features (X) e target (y)**
Objetivo: o perfil físico (idade, altura, peso, sexo) do atleta consegue discriminar o esporte/evento que ele/ela compete?
Como percebi que tenho valores faltantes nos atributos idade, altura e peso, antes de dividir os dados eu vou remover as linhas com valores ausentes. Percebi também que temos uma lista grande de esportes, optando assim por usar um conjunto menor de dados para modelagem. Para promover melhor balanceamento, criei a coluna esporte_modalidade.

````python
# Separar features (X) e target (y).
# Objetivo: o perfil físico (idade, altura, peso, sexo) do atleta consegue discriminar o esporte_modalidade que ele/ela compete?
# Como percebi que tenho valores faltantes nos atributos idade, altura e peso, antes de dividir os dados
# eu vou remover as linhas com valores ausentes. Percebi tb que temos uma lista grande de esportes, optando assim por
# usar um conjunto menor de dados para modelagem.

limite = 500

cols_modelo = ['idade', 'altura', 'peso', 'sexo_bin', 'esporte','evento']
df_modelo = df[cols_modelo].dropna().copy()
df_modelo['esporte_modalidade'] = df_modelo['esporte'] + " / " + df_modelo['evento']

# filtra esportes_modalidade com > (limite) ocorrências
modelo_filtrado = df_modelo['esporte_modalidade'].value_counts()
modelo_filtrado = modelo_filtrado[modelo_filtrado > limite].index
df_modelo = df_modelo[df_modelo['esporte_modalidade'].isin(modelo_filtrado)].copy()

# separa X e y
X = df_modelo[['idade', 'altura', 'peso', 'sexo_bin']]
y = df_modelo['esporte_modalidade']
````

````python
print("Tamanho final:", df_modelo.shape)
print("\nTop esportes (pós-filtro):")
print(df_modelo['esporte'].value_counts().head(20))

print("\nHead esporte/modalidade (target):")
print(y.value_counts().head(10))

print("\nTail esporte/modalidade (target):")
print(y.value_counts().tail(10))
````

````python
# Dividir os dados em conjuntos de treino e teste
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)
````
